```python
#pip install opencv-python
from google.colab.patches import cv2_imshow
import cv2

#Load an image
image = cv2.imread('/content/download.jpg')

#Display Image
cv2_imshow(image)
print('Image Dimension:',image.shape)
```

```python
import matplotlib.pyplot as plt

#Resize Image
imageResized= cv2.resize(image, (300,300))
cv2_imshow(imageResized)
```

```python
#Bigger, Half & stretch
half = cv2.resize(imageResized, (0,0), fx =0.5,fy=0.5)
cv2_imshow(half)
```

```python
bigger = cv2.resize (imageResized,(1080,1080))
cv2_imshow(bigger)
```

```python
stretch = cv2.resize(imageResized, (300,60), interpolation = cv2.INTER_LINEAR)
cv2_imshow(stretch)
```

```python
# Subplot Image
titles = ["Original","Half",  "Bigger", "Interpolation Neareast"]
images= [image,imageResized, bigger, stretch]
count = 4
for i in range(count):
    plt.subplot(4,4,i+1)
    plt.title(titles[i])
    plt.imshow(images[i])

plt.show()

```

```python
#Rotation of Image
(h,w) = image.shape[:2]
center = (w/2,h/2)
angle90 = 90
angle180 = 180
angle270 = 270
scale =1.0

rotate_90 = cv2.getRotationMatrix2D(center, angle90, scale)
rotate90 = cv2.warpAffine(image, rotate_90, (h,w))
cv2_imshow(rotate90 )
```

```python
# Chaning the color
chnageColor=cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
cv2_imshow(chnageColor)
```
